```python
#plot a graph where x varies frm 0 to 5 , y=x+3
#give x label as 'x-label' , y label as 'y-label', title as 'y=x+3'

x=np.array([0,1,2,3,4,5])
y=x+3
plt.title('y=x+3')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.plot(x,y)

plt.grid()
plt.show()
```
